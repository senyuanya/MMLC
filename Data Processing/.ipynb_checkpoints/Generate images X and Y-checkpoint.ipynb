```python
def Data(path_num):
    path = all_file_list[path_num]
    data = pd.read_csv('./indexdata/'+ path)
    stk_data = data.iloc[::-1].reset_index(drop=True)
    return stk_data

def getDaliyVol(data,span0=100):
    data['Return'] = data['Close']/data['Close'].shift(1)-1
    sigma = data['Return'].ewm(span=span0).std()
    return sigma

def TBL(df,events,width):
    result = events[['Date','VB']].copy(deep=True)
    if width[0] > 0:
        events['UB'] = width[0]*events['Vol']
    else:
        events['UB'] = np.nan
    if width[1] > 0:
        events['DB'] = -width[1]*events['Vol']
    else:
        events['DB'] = np.nan
    for col, date, vb in result.itertuples( ):
        df0 = df[(pd.to_datetime(df['Date'])>date) & (pd.to_datetime(df['Date'])<vb)].copy(deep = True)
        df0['Return'] = df0['Close']/df.loc[df['Date']==date,'Close'].iloc[0]-1
        
        idx = (result['Date']==date)
        
        result.loc[idx,'ut']=pd.to_datetime(df0.loc[df0['Return'] > events.loc[idx,'UB'].iloc[0],'Date'].min())
        result.loc[idx,'dt']=pd.to_datetime(df0.loc[df0['Return'] < events.loc[idx,'DB'].iloc[0],'Date'].min())
        
    return result

def get_first_touch(df, events,width):
    result = TBL(df,events,width)
    result['First']= result[['VB','ut','dt']].dropna(how='all').min(axis=1)
    result['First']= result['First'].astype(str)
    return result

def get_labels(result,data):
    out = result.copy(deep=True)
    touch = result[['VB','ut','dt','First']]
    for i in range(len(touch)):
        if pd.to_datetime(touch.loc[i,'First']) == touch.loc[i,'VB']:
            out.loc[i,'label'] = 1
        elif pd.to_datetime(touch.loc[i,'First']) == touch.loc[i,'ut']:
            out.loc[i,'label'] = 2
        else:
            out.loc[i,'label'] = 0
    outcome=pd.merge(data,out, on=['Date'], how='left')
    return outcome
```

```python
def datasplit(stk_data,sequence_length,window_size,percent_training,percent_validing):
    """
    sequence_length: Number of x_i for each y_i
    window_size: Number of y_i 
    """
    # Define the training set`
#     percent_training: float = 0.60
    num_training_samples = int(np.ceil(percent_training*len(stk_data))) # number of training samples in int
    training_set = stk_data.iloc[:num_training_samples, 4:5].values # Extract the Close values for training data
    training_label = stk_data.iloc[:num_training_samples, 12:13].values 
    # Define the validation set
#     percent_validing: float = 0.20
    num_validing_samples = int(np.ceil(percent_validing*len(stk_data))) # number of validing samples in int    
    validing_set = stk_data.iloc[num_training_samples:(num_training_samples + num_validing_samples), 4:5].values # Extract the Close values for validing data
    validing_label = stk_data.iloc[num_training_samples:(num_training_samples + num_validing_samples), 12:13].values
    #Define the test data
    num_testing_samples: int = len(stk_data) - (num_training_samples + num_validing_samples) # number of test samples in int 
    testing_set = stk_data.iloc[-num_testing_samples:, 4:5].values # Extract the Close values for test data
    testing_label = stk_data.iloc[-num_testing_samples:, 12:13].values
    print('Training set shape:', training_set.shape, '\n', 'Validinging set shape:', validing_set.shape, '\n', 'Testing set shape:', testing_set.shape)
    ####Train
#     sequence_length: int = 22 # Full Number of x_i for each y_i
#     window_size: int = 10   #Number of y_i (short)
    x_train = np.array([training_set[i - sequence_length:i, :] for i in range(sequence_length, len(training_set))]).squeeze(axis=2)
    x_train = np.delete(x_train,np.s_[-window_size:],axis = 0)
    y_train = np.array([training_set[i-window_size:i,:] for i in range(sequence_length+window_size, len(training_set))]).squeeze(axis=2)# 2D (num_training_samples, 1)
    label_train = np.array([training_label[i,:] for i in range(sequence_length,len(training_set))]).squeeze(axis=1)
    label_train = np.delete(label_train,np.s_[-window_size:],axis = 0)
    ####Valid
    x_valid= np.array([validing_set[i - sequence_length:i, :] for i in range(sequence_length, len(validing_set))]).squeeze(axis=2)
    x_valid = np.delete(x_valid,np.s_[-window_size:],axis = 0)
    y_valid = np.array([validing_set[i-window_size:i,:] for i in range(sequence_length+window_size, len(validing_set))]).squeeze(axis=2)# 2D (num_training_samples, 1)
    label_valid = np.array([validing_label[i,:] for i in range(sequence_length,len(validing_set))]).squeeze(axis=1)
    label_valid = np.delete(label_valid,np.s_[-window_size:],axis = 0)
    ###Test
    x_test = np.array([testing_set[i - sequence_length:i, :] for i in range(sequence_length, len(testing_set))]).squeeze(axis=2)
    x_test = np.delete(x_test,np.s_[-window_size:],axis = 0)
    y_test = np.array([testing_set[i-window_size:i,:] for i in range(sequence_length+window_size, len(testing_set))]).squeeze(axis=2)# 2D (num_training_samples, 1)
    label_test = np.array([testing_label[i,:] for i in range(sequence_length,len(testing_set))]).squeeze(axis=1)
    label_test = np.delete(label_test,np.s_[-window_size:],axis = 0)
    print('x_train,y_train,label_train,x_valid,y_valid,label_valid,x_test,y_test,label_test:',x_train.shape,y_train.shape,label_train.shape,x_valid.shape,y_valid.shape,label_valid.shape,x_test.shape,y_test.shape,label_test.shape)
    return x_train,y_train,label_train,x_valid,y_valid,label_valid,x_test,y_test,label_test
```

```python
def distr_minmaxY(path_num,window_size,y_train,num):#or y_valid,num:0:train/1:valid
    a = ['train','valid','test']
    path_label = "./dataset/pre"+str(window_size)+"/"+str(path_num)+"/"
    baseline = 0.005*int(window_size)*2
    tmp = pd.DataFrame(np.tile(y_train[:, 0].reshape(-1,1), (1, window_size)))
    y_price = y_train / tmp - 1 # scaling
    y_price['max_val']= y_price.max(axis=1) #取出该最大值
    y_price['max_idx'] = y_price.idxmax(axis=1) #求一行的最大值对应的索引
    y_max = y_price[['max_idx','max_val']]
    y_price= y_price.drop(columns=['max_idx','max_val'])
    y_price['min_val'] = y_price.min(axis=1)
    y_price['min_idx'] = y_price.idxmin(axis=1)
    y_min = y_price[['min_idx','min_val']]
    plt.figure(figsize = (8,4))#设置figure的尺寸
    x= np.arange(len(y_min))
    y1 = np.array(y_max['max_val'])
    y2 = np.array(y_min['min_val'])
    plt.axhline(y=baseline,label='baseline', color='black', linestyle='-')#0.5%*10（pred10）*2
    plt.axhline(y=-baseline, color='black', linestyle='-')
    plt.plot(x,y1,label="$max(rate)$",color="red",linewidth=2)#设置格式
    plt.plot(x,y2,label="$min(rate)$",color="blue",linewidth=2)#设置格式
    plt.legend(loc='best')
    plt.savefig(path_label+'{}.jpg'.format(a[num]), dpi=500, bbox_inches='tight')
    plt.show()
```

```python
def convert_to_gaf(X):
    gaf = GramianAngularField()
    X_images = gaf.fit_transform(X)
    return X_images

def convert_to_RP(X):
    rp = RecurrencePlot(threshold='point', percentage=20)
    X_images = rp.fit_transform(X)
    return X_images

def convert_to_RPgaf(X):
    gaf_X = convert_to_gaf(X)
    rp_X = convert_to_RP(X)
    mg_mix = cv2.addWeighted(gaf_X, 0.5, rp_X, 0.5, 0)
    return mg_mix

def weight_matrix(y):
    x= np.zeros((y.shape[0]))
    for k in range(y.shape[0]):
        a = y[k][-1]-y[k][0]
        if a <0:
            x[k]=-1
        else:
            x[k]=1
    return x
def convert_to_weightgaf(X):
    gaf_X = convert_to_gaf(X)
    weight=weight_matrix(X)[:,np.newaxis,np.newaxis]
    mg_mix = weight*gaf_X
    return mg_mix

def convert_to_weightRP(X):
    RP_X = convert_to_RP(X)
    weight=weight_matrix(X)[:,np.newaxis,np.newaxis]
    mg_mix = weight*RP_X
    return mg_mix
```

```python
def plt_imshowX(title, image):
    plt.imshow(image)
    plt.title(title)
    plt.show()
```

```python
def save_imagesX(arr, path, code):#'000000';'100000';'200000000'; '300000000000';'400000000000000'
    j = 0
    pic_name = []
    for i in arr:
        plt.imsave(path + "/" + code + str(j) + ".png", i)
        save_filename = code +'%d.png'%j
        j += 1
        pic_name.append(save_filename)
    return pic_name
```

```python
def saveimages(path_num,window_size,code,x_train,x_valid,x_test):
    path_label = "./dataset/pre"+str(window_size)+"/"+str(path_num) 
    print("Converting to weightgaf images")
    X_train_gaf = convert_to_weightRP(x_train)
    X_valid_gaf = convert_to_weightRP(x_valid)
    X_test_gaf = convert_to_weightRP(x_test)
    plt_imshowX("X_train_gaf[0]", X_train_gaf[0])
    plt_imshowX("X_valid_gaf[0]", X_valid_gaf[0])
    plt_imshowX("X_test_gaf[0]", X_test_gaf[0])
    
    print("Saving weightgaf images")
    train_pic_name = save_imagesX(X_train_gaf, path_label + "/weightrpImages/TrainX", code)
    valid_pic_name = save_imagesX(X_valid_gaf, path_label + "/weightrpImages/ValidX", code)
    test_pic_name = save_imagesX(X_test_gaf, path_label + "/weightrpImages/TestX", code)
```

```python
path_num = 2##8,34,35,42,49
window_size = 15
width=[2.0,2.0] #pre10:1.5,pre13:1.8,pre15:2.0
code = '200000000' #'000000';'100000';'200000000'; '300000000000';'400000000000000'
sequence_length=30
```

```python
data =Data(path_num)#8,34,35,42,49
events = data[['Date']].copy(deep=True)
events['VB'] = pd.to_datetime(events['Date'])+pd.Timedelta(days=window_size)
events['Vol'] = getDaliyVol(data)
events.head(3).append(events.tail(3))
```

```python
result = get_first_touch( data, 
                          events,
                          width )
result.head(20).append(result.tail(20))
```

```python
outcome = get_labels(result,data)
outcome.drop(outcome.tail(window_size).index,inplace=True) # 从尾部去掉n行
outcome
```

```python
x_train,y_train,label_train,x_valid,y_valid,label_valid,x_test,y_test,label_test = datasplit(outcome,sequence_length=sequence_length,window_size=window_size,percent_training=0.6,percent_validing=0.2)
```
